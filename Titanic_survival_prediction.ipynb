#Importing Libraries


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Importing Dataset

```python
dataset = pd.read_csv('/content/drive/MyDrive/CODSOFT/TASK 1/tested.csv')
```

```python
column_list = ["PassengerId", "Pclass",	"Name",	"Sex",	"Age",	"SibSp",	"Parch",	"Ticket",	"Fare",	"Cabin", "Embarked", "Survived"]
dataset = dataset.reindex(columns = column_list)
```

```python
dataset.drop(['PassengerId', 'Name',	'Ticket',	'Cabin'], axis=1,inplace=True)
```

```python
pd.isnull(dataset).sum()
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.info()
```

#Data Preprocessing

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le6 = LabelEncoder()
X[:,1] = le1.fit_transform(X[:,1])
X[:,6] = le6.fit_transform(X[:,6])
```

```python
print(X)
```

#Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.33,random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
print(X_train)
```

#Training Dataset

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier()
model_knn.fit(X_train,Y_train)
```

#Model predictions


```python
from sklearn.metrics import classification_report
```

```python
y_pred_knn = model_knn.predict(X_test)
print(classification_report(Y_test,y_pred_knn))
```
